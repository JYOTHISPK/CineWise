```python
import pandas as pd
from pathlib import Path
```

```python
PROJECT_PATH = Path.cwd()
while not (PROJECT_PATH / "data").exists() :
    PROJECT_PATH = PROJECT_PATH.parent
```

```python
df=pd.read_csv(PROJECT_PATH/ "data" / "cleaned movies.csv" )
```

```python
df.info()
```

```python
print(df.isnull().sum())
```

```python
df.head()
```

```python
def get_movie_era(Year) :

    if Year < 1980 :
        return "before_1980"
    elif Year < 1990 :
        return "1980_1990"
    elif Year < 2000 :
        return "1990_2000"
    return "after_2000"
```

```python
df['Era']=df['Year'].apply(get_movie_era)
```

```python
df["Genre"]=df["Genre"].str.replace(",","")
```

```python
df['Combined_Features']=df["Genre"] + " " + df["Language"]+ " " + df["Era"]
```

```python
(df["Genre"] == "unknown").sum()
```

```python
df = df[df["Genre"] != "unknown"]
```

```python
df = df.reset_index(drop=True)
```

```python
df.to_csv(PROJECT_PATH/ "data" / "more cleaned movies.csv" )
```
